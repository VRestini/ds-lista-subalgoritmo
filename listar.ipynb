```python
print("--- MENU DE EXERCÍCIOS ---")
print(" EXERCÍCIO 1: PREENCHER UMA LISTA \n EXERCÍCIO 2: EXIBE A LISTA \n EXERCÍCIO 3: CONTA A QUANTIDADE DE ELEMENTOS \n EXERCÍCIO 4: RETORNA O PRIMEIRO ÍNDICE DO VALOR PASSADO COMO PARÂMETRO \n EXERCÍCIO 5: IRÁ CONTAR A QUANTIDADE EXATA DE ELEMENTOS EXISTENTES NA LISTA ATRAVÉS DO VALOR PASSADO COMO PARÂMETRO \n EXERCÍCIO 6: CONTA QUANTOS ELEMENTOS INTEIROS EXISTEM NA LISTA \n EXERCÍCIO 7: CONTA QUANTOS ELEMENTOS STRINGS EXISTEM NA LISTA \n EXERCÍCIO 8: CONTA QUANTOS ELEMENTOS LÓGICOS EXISTEM NA LISTA \n EXERCÍCIO 9: CONTA QUANTOS ELEMENTOS FLOAT EXISTEM NA LISTA \n EXERCÍCIO 10: VERIFICA TODOS ELEMENTOS DE UMA LISTA E COPIA OS INTEIROS PARA OUTRA \n" )
lista = []
lista2 = []
elemento = "0"
def preenche_lista(lista: list) -> None:
    elemn = 's'
    while elemn != "=":
        elemn = input("Digite o elemento que será inserido, se você desejar não inserir mais valores basta digitar [=]")
        lista.append(elemn)
        if elemn == "=":
            lista.remove("=")
        
  
def exibe_lista(lista : list) -> None:
    print(lista)

def conta_elemento(lista: list) -> int:
    x = len(lista)
    return x

def retorna_indice(elemento: str) -> int:
    elemento = input()
    x = lista.index(elemento)
    return x 

def busca(lista: list, elemento: str) -> int:
    elemento = input()
    x = lista.count(elemento)
    return x

def conta_inteiro(lista: list) -> int:
    qtd = 0
    for elem in lista:
        if elem.isdigit():
            qtd += 1
    return qtd

def conta_string(lista:list) -> int:
    qtd = len(lista)
    for elem in lista:
        if elem.isdigit():
            qtd = qtd - 1
        elif elem == "True" or elem == "False":
            qtd = qtd - 1
    return qtd

def conta_boolean(lista:list) -> int:
    qtd = 0
    for elem in lista:
        if elem == "True" or elem == "False":
            qtd =+ 1
    return qtd      

def conta_float(lista:list) -> int:
    qtd = 0
    for elem in lista:
        if elem.replace('.', '',1).isdigit():
            qtd =+ 1
    return qtd
def copia_int(lista:list, lista2:list) -> None:
    for elem in lista:
        if elem.isdigit():
            lista2.append(int(elem))
    print("Lista 2:" , lista2)            



cont = 's'
while cont == 's' or cont == 'S':
    op = int(input("Escolha seu exercício:"))
    match op:
        case 1:
            preenche_lista(lista)
        case 2:
            exibe_lista(lista)
        case 3:
            print(conta_elemento(lista))
        case 4:
            print(retorna_indice(elemento))
        case 5:
            print(busca(lista, elemento))
        case 6:
            print(conta_inteiro(lista))
        case 7:
            print (conta_string(lista))
        case 8:
            print (conta_boolean(lista))
        case 9:
            print (conta_float(lista))
        case 10:
            copia_int(lista, lista2)
        case _:
            print("Exercício inexistente")

    cont = input("Deseja continuar no programa? Digite [s] para sim e qualquer outro valor para não")
```
